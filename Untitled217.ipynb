```python
import torch
import torch.nn as nn
import pandas as pd
import imageio
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
from tqdm import tqdm
```

```python
model = torch.hub.load('pytorch/vision:v0.10.0', 'resnet18', pretrained=True)
```

```python
model.eval()
```

```python
model.fc = nn.Linear(512,5)
```

```python
class MultiInputModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.ResNetModel = model
        self.LinBlock1 = nn.Linear(4,64)
        self.ReLU1 = nn.ReLU(inplace=True)
        self.LinBlock2 = nn.Linear(64,32)
        self.ReLU2 = nn.ReLU(inplace=True)
        self.Dropout1 = nn.Dropout(p=0.3)
        self.LinBlock3 = nn.Linear(32,5)
        self.LinBlockFinal = nn.Linear(10,1)
    def forward(self,img,num):
        img = self.ResNetModel(img)
        num = self.LinBlock1(num)
        num = self.ReLU1(num)
        num = self.LinBlock2(num)
        num = self.ReLU2(num)
        num = self.Dropout1(num)
        num = self.LinBlock3(num)
        print(img.shape)
        img = img.flatten()
        print(img.shape)
        print(num.shape)
        final = torch.concat([img,num])
        final = self.LinBlockFinal(final)
        return final
```

```python
modelinstance = MultiInputModel()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.RMSprop(modelinstance.parameters(),lr=0.01)
```

```python
def training_loop(model,optimizer,criterion,dataloader):
    for batch,(img,num,y) in enumerate(dataloader):
        y_pred = model(img,num)
        loss = criterion(y_pred,y);
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
